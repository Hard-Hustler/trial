## Exercise 1

Consider a neural network where, for each scalar input `x`, it outputs a predicted value `yhat` as follows:

    zh = wh*x + bh
    uh = 1/(1 + exp(-zh))   # Sigmoid activation
    zo = uh.dot(wo) + bo
    yhat = zo               # No activation

Using the parameter values below, for scalar inputs `x` in the range of -4 to 8:

*  Plot `uh` vs `x`.  Since there are three hidden units, your graph should have three curves
*  Plot `yhat` vs `x`.  Since there is one output unit, your graph should have one curve

```python
x = np.linspace(-4,8,100)
wh = np.array([1,1,1])
bh = -np.array([0,2,4])
wo = np.array([1,-2,0.5])
bo = 0.1

# TODO
#   uh = ...
#   yhat = ...

# Hidden layer pre-activation
zh = np.outer(x, wh) + bh    # shape (100, 3)

# Sigmoid activation
uh = 1 / (1 + np.exp(-zh))

# Output layer (linear, no activation)
yhat = uh @ wo + bo

plt.figure(figsize=(8,5))
for i in range(3):
    plt.plot(x, uh[:, i], label=f'u_h{i+1}')
plt.title('Hidden Layer Activations (u_h) vs x')
plt.xlabel('x')
plt.ylabel('u_h')
plt.legend()
plt.grid(True)
plt.show()

# plot 2
plt.figure(figsize=(8,5))
plt.plot(x, yhat, color='black', linewidth=2)
plt.title('Predicted Output (ŷ) vs x')
plt.xlabel('x')
plt.ylabel('ŷ')
plt.grid(True)
plt.show()

```
